### Imports

````python
import cv2
import matplotlib.pyplot as plt
````

### Reading and inspecting the image

<img src="sample_images/img_rep.png"/>

````python
img_path = 'sample_images/machine-learning-meetup.png'
````

````python
img = cv2.imread(img_path)
````

### Verifying shape (width, height, channels)

````python
print(img.shape)
````

### Verifying data type

````python
print(img.dtype)
````

### Showing the image
Note the colors are not correct. This is because OpenCV loads images in BGR by default.

````python
plt.imshow(img)
````

### Changing the channel order

````python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
````

````python
plt.imshow(img_rgb)
````

### Assessing difference of BGR and RGB 

````python
print('BGR')
print('B:', round(img[:,:,0].mean(), 2))
print('G:', round(img[:,:,1].mean(), 2))
print('R:', round(img[:,:,2].mean(), 2))
````

````python
print('RGB')
print('R:', round(img_rgb[:,:,0].mean(), 2))
print('G:', round(img_rgb[:,:,1].mean(), 2))
print('B:', round(img_rgb[:,:,2].mean(), 2))
````

### Resizing the image

````python
img_rgb_resized = cv2.resize(img_rgb, (300, 300))
````

````python
plt.imshow(img_rgb_resized)
````

````python
print(img_rgb_resized.shape)
````

### Converting to grayscale

````python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
````

````python
print(img_gray.shape)
````

````python
plt.imshow(img_gray, cmap='gray')
````
