### Imports

````python
from skimage import io
from skimage import transform
from skimage import img_as_ubyte

from skimage.color import rgb2gray
from skimage.color import rgba2rgb

import matplotlib.pyplot as plt
````

### Reading and inspecting the image

<img src="sample_images/img_rep.png"/>

````python
img_path = 'sample_images/machine-learning-meetup.png'
````

````python
img = io.imread(img_path)
````

### Verifying shape (width, height, channels)
Note that we now have 4 channels (RGBA)

````python
print(img.shape)
````

### Verifying data type

````python
print(img.dtype)
````

### Showing the image
Note the colors are correct. This is because the image is loaded as RGBA.

````python
plt.imshow(img)
````

### Converting to RGB

````python
img_rgb = rgba2rgb(img)
````

````python
plt.imshow(img_rgb)
````

````python
print(img_rgb.shape)
````

Notice the data type changed to float64. Does it make a difference?

````python
print(img_rgb.dtype)
````

If we look at the mean of the pixels, we can see an important difference. Remember OpenCV values?

````python
print('RGB - float64')
print('R:', round(img_rgb[:,:,0].mean(), 2))
print('G:', round(img_rgb[:,:,1].mean(), 2))
print('B:', round(img_rgb[:,:,2].mean(), 2))
````

### Converting to uint8
After converting to uint8, we can see the same results of OpenCV.

````python
img_rgb_uint8 = img_as_ubyte(img)
````

````python
print('RGB - uint8')
print('R:', round(img_rgb_uint8[:,:,0].mean(), 2))
print('G:', round(img_rgb_uint8[:,:,1].mean(), 2))
print('B:', round(img_rgb_uint8[:,:,2].mean(), 2))
````

````python
plt.imshow(img_rgb_uint8)
````

### Resizing the image

````python
img_rgb_resized = transform.resize(img_rgb, (300, 300))
````

````python
plt.imshow(img_rgb_resized)
````

````python
print(img_rgb_resized.shape)
````

### Converting to grayscale

````python
img_gray = rgb2gray(img_rgb)
````

````python
print(img_gray.shape)
````

````python
plt.imshow(img_gray, cmap='gray')
````

````python
plt.imshow(img_gray2, cmap='gray')
````
